Considere as seguintes frequências digitais (em rad): 0,1 0,4  pi/10 pi/5 .

No gráfico da esquerda, plote as 4 senoides com 100 amostras.

No gráfico da direita, plote o valor absoluto da DFT dessas 4 senoides.

Comente o que foi observado.

obs. São apenas dois gráficos, cada um com quatro curvas.

```python
from numpy import *
from scipy import *
from matplotlib.pyplot import *
```

```python
t = arange(0, 22*pi, (22*pi)/100)
t_eixo = arange(0, 100, 1)
sen1 = sin(t*0.1)
sen2 = sin(t*0.4)
sen3 = sin(t*(pi/10))
sen4 = sin(t*(pi/5))
```

```python
fig, axs = subplots(1,2,figsize=(20,5))
axs[0].plot(t_eixo,sen1, label="w = 0.1")
axs[0].plot(t_eixo,sen2, label="w = 0.4")
axs[0].plot(t_eixo,sen3, label="w = pi/10")
axs[0].plot(t_eixo,sen4, label="w = pi/5")
axs[1].plot(t_eixo,fft(sen1), label="w = 0.1")
axs[1].plot(t_eixo,fft(sen2), label="w = 0.4")
axs[1].plot(t_eixo,fft(sen3), label="w = pi/10")
axs[1].plot(t_eixo,fft(sen4), label="w = pi/5")
legend()
show()
```
