```python
from numpy import *
from scipy.signal import *
from matplotlib.pyplot import *
```

```python
t = arange(0,16*pi, (16*pi)/64)
freq = fft.fftfreq(t.shape[-1])
a = sin(t)

b = zeros(64)
b[16:48] = 1

c = a*b

d = a[0:32]

t2 = arange(0,8*pi,(8*pi)/64)
d2 = sin(t2*2)
freq2 = fft.fftfreq(t2.shape[-1])
```

```python
plot(a, label='a')
plot(b, label='b')
plot(c,'-.', label='c')
plot(d,'--', label='d')
legend()
show()
```

```python
plot(freq,abs(fft.fft(a)), label='a')
plot(freq,abs(fft.fft(b)), label='b')
plot(freq,abs(fft.fft(c)), label='c')
plot(freq[16:48],abs(fft.fft(d)), label='d')
legend()
show()
```

O que é notável no gráfico é a diferença dos gráficos de "c" e "d". 

Percebe-se que ambos possuem a mesma amplitude já que possuem o mesmo número de ciclos.

A frequência de "d" está fora do esperado, já que teoricamente esse vetor tem a mesma frequência dos vetores "a" e "c".
