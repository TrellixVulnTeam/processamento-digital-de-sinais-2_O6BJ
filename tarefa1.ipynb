```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
```

### 1) Gerar os seguintes sinais e visualizá-los (comandos plot e stem). Gerar 200 valores para n (de 0 a 199).
### a) senoide com frequência w=0,2 rad ( sin(w*n) )

```python
a = np.linspace(0, 199, 200)

plt.figure(figsize=(8,4.5))
plt.plot(a, np.sin(a*0.2))
```

### b) senoide com 8 ciclos

```python
b = np.linspace(0, 16*np.pi, 200)

plt.figure(figsize=(8,4.5))
plt.plot(b, np.sin(b))
```

### c) cosenoide com 3 ciclos

```python
c = np.linspace(0, 6*np.pi, 200)

plt.figure(figsize=(8,4.5))
plt.plot(c, np.cos(c))
plt.show()
```

### d) senoide de 60 Hz (frequência de amostragem = 800 amostas por segundo = 800 Hz)

https://thenucleargeeks.com/2019/09/28/dsp-icebreaker-drawing-a-sine-wave-in-python-of-100-hz/

```python
f_d = 60
fs_d = 800
w_d = 2 * np.pi * f_d / fs_d

t_d = np.linspace(0, 199, 200)
amp_d = np.sin(w_d * t_d)

plt.figure(figsize=(16,4.5))
plt.stem(t_d , amp_d)
```

### e) exponencial complexa y=exp(j*w*n), calcular w para 3 ciclos. Plotar parte real e a parte imaginária

$ Z = a+jb = M \angle \theta = M e^{j \theta} = M (cos \theta + j sen \theta)$

$ M = \sqrt{a^2 + b^2} $

$ \theta = arctg(b/a)$

$ a = M cos \theta $

$ b = M sen \theta $

```python
e = np.linspace(0, 6*np.pi, 200)
ee = 1j*e
ee = np.exp(ee)

plt.figure(figsize=(8,5))
plt.plot(np.real(ee), label="Real")
plt.plot(np.imag(ee), label="Imag.")
plt.legend()
```

### f) exponencial oscilante crescente. y=alfa.^n. Escolher valor complexo de alfa. Plotar parte real

```python
n_f  = np.linspace(0 , 199 , 200)
alfa_f = .35+1j
print(np.abs(alfa_f))
f = alfa_f ** n_f

plt.figure(figsize=(8,4.5))
plt.plot(np.real(f))
```

### g) exponencial oscilante decrescente. y=alfa.^n. Escolher  valor complexo de alfa. Plotar parte real

```python
n_g  = np.linspace(0 , 199 , 200)
alfa_g = .5+.8j
print(abs(alfa_g))
g = alfa_g ** n_g

plt.figure(figsize=(8,4.5))
plt.plot(np.real(g))
```
