### d) senoide de 60 Hz (frequência de amostragem = 800 amostas por segundo = 800 Hz)

https://thenucleargeeks.com/2019/09/28/dsp-icebreaker-drawing-a-sine-wave-in-python-of-100-hz/

```python
f_d = 60
fs_d = 800
w_d = 2 * np.pi * f_d / fs_d

t_d = np.linspace(0, 200, 200)
amp_d = np.sin(w_d * t_d)

plt.plot(t_d , amp_d)
```

### e) exponencial complexa y=exp(j*w*n), calcular w para 3 ciclos. Plotar parte real e a parte imaginária

$ Z = a+jb = M \angle \theta = M e^{j \theta} = M (cos \theta + j sen \theta)$

$ M = \sqrt{a^2 + b^2} $

$ \theta = arctg(b/a)$

$ a = M cos \theta $

$ b = M sen \theta $

```python
x = np.linspace(0, 6*np.pi, 200)
xx = x + 1j * x[:, np.newaxis]
out = np.exp(xx)
plt.plot(np.angle(out))
```
